```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/heart.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x = df.drop(['target'],axis=1)
x
```

```python
y = df['target']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=0)
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.svm import SVC
svm = SVC(gamma ='auto')
svm = svm.fit(xtrain,ytrain)
svm
```

```python
svm = svm.score(xtest,ytest)
svm
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn = knn.fit(xtrain,ytrain)
knn
```

```python
knn = knn.score(xtest,ytest)
knn
```

```python
from sklearn.linear_model import LogisticRegression
Logistic = LogisticRegression()
Logistic = Logistic.fit(xtrain,ytrain)
Logistic
```

```python
Logistic= Logistic.predict(xtest)
Logistic
```

```python
from sklearn.ensemble import RandomForestClassifier
random = RandomForestClassifier()
random = random.fit(xtrain,ytrain)
random
```

```python
random = random.score(xtest,ytest)
random
```

```python
from sklearn.tree import DecisionTreeClassifier
decision = DecisionTreeClassifier()
decision = decision.fit(xtrain,ytrain)
decision
```

```python
decision = decision.score(xtest,ytest)
decision
```
