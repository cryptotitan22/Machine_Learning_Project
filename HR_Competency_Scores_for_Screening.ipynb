```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/dataset.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x = df.drop(['call_for_interview'],axis=1)
x
```

```python
y = df['call_for_interview']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
from sklearn.svm import SVC
svm = SVC(gamma='auto')
svm = svm.fit(xtrain,ytrain)
svm
```

```python
svm = svm.score(xtest,ytest)
svm
```

```python
from sklearn.ensemble import RandomForestClassifier
random = RandomForestClassifier()
random = random.fit(xtrain,ytrain)
random
```

```python
random = random.score(xtest,ytest)
random
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree = tree.fit(xtrain,ytrain)
tree
```

```python
tree = tree.score(xtest,ytest)
tree
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn = knn.fit(xtrain,ytrain)
knn
```

```python
knn = knn.score(xtest,ytest)
knn
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic = logistic.fit(xtrain,ytrain)
logistic
```

```python
logistic = logistic.score(xtest,ytest)
logistic
```
